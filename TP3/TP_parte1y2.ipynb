```python
import pandas as pd

# Cargamos el archivo Excel en un DataFrame
df1 = pd.read_excel('C:/Users/Usuario/Documents/Ciencia de datos/TP3 2/usu_individual_T124.xlsx')
# Mostramos las primeras filas del DataFramez
df1.head()
```

```python
# Cargamos el archivo Excel en un DataFrame
df2 = pd.read_stata('C:/Users/Usuario/Documents/Ciencia de datos/TP3 2/Individual_t104.dta')
# Mostramos las primeras filas del DataFrame
df2.head()
```

```python
# Convertimos los nombres de las columnas a minúsculas para df1
df1.columns = df1.columns.str.lower()

# Convertimos los nombres de las columnas a minúsculas para df2 (por las dudas de que haya alguno en mayúscula y no lo estemos viendo)
df2.columns = df2.columns.str.lower()

# Mostramos las primeras filas de los DataFrames para verificar los cambios
df1.head(), df2.head()
```

```python
# Obtenemos las columnas de cada DataFrame
columnas_df1 = set(df1.columns)
columnas_df2 = set(df2.columns)

# Columnas comunes
columnas_comunes = columnas_df1.intersection(columnas_df2)

# Columnas diferentes
columnas_diferentes_df1 = columnas_df1.difference(columnas_df2)
columnas_diferentes_df2 = columnas_df2.difference(columnas_df1)

# Mostramos resultados
print("Columnas comunes:")
print(columnas_comunes)

print("\nColumnas en df1 pero no en df2:")
print(columnas_diferentes_df1)

print("\nColumnas en df2 pero no en df1:")
print(columnas_diferentes_df2)
```

```python
#A
# Unimos ambos DataFrames uno debajo del otro
df_combined = pd.concat([df1, df2], ignore_index=True)

# Mostramos el DataFrame combinado para verificar
df_combined.head()
```

```python
#A (seguimos)

# Filtramos el DataFrame combinado para mantener solo las filas con los valores especificados en la columna 'aglomerado'
df_filtered = df_combined[df_combined['aglomerado'].isin(['Partidos del GBA', 'Ciudad de Buenos Aires', 33, 32])]

# Mostramos el DataFrame filtrado para verificar
df_filtered.head()
```

```python
#B
# Seleccionamos las columnas con las que nos queremos quedar
columnas_seleccionadas = ['ano4','aglomerado', 'ch06', 'ch04', 'nivel_ed', 'estado', 'ipcf', 'cat_inac', 'ch08','ch07']

# Creamos el nuevo DataFrame
df_nuevo = df_filtered[columnas_seleccionadas]

df_nuevo.head()

```

```python
print(df_nuevo['ano4'].dtype)

df_nuevo['ano4'] = df_nuevo['ano4'].astype(int)

df_nuevo['ipcf'] = df_nuevo['ipcf'].astype(int)

print(df_nuevo['ano4'].dtype)

print(df_nuevo['ipcf'].dtype)

df_nuevo.head()
```

```python
#Vamos a cambiar los numeros de la base de datos de 2024 por los nombres (como aparecen en 2004)
# Creamos un diccionario de mapeo para las columnas que necesitan ser cambiadas
mapeo_ch04 = {1: 'Varón', 2: 'Mujer'}
mapeo_nivel_ed = {1: 'Primaria Incompleta (incluye educación especial)', 2: 'Primaria Completa', 3: 'Secundaria Incompleta', 4: 'Secundaria Completa', 5: 'Superior Universitaria Incompleta', 6: 'Superior Universitaria Completa', 7: 'Sin instrucción', 9: 'NS/NC'}
mapeo_estado = {0: 'Entrevista individual no realizada (no respuesta al cuestionario individual)', 1: 'Ocupado', 2: 'Desocupado', 3:'Inactivo', 4:'Menor de 10 años'}
mapeo_aglomerado = {32: 'Ciudad de Buenos Aires', 33:'Partidos del GBA'}
mapeo_cat_inac= {1: 'Jubilado/pensionado', 2: 'Rentista', 3:'Estudiante', 4:'Ama de casa', 5:'Menor de 10 años', 6:'Discapacitado', 9:'Otros'}
mapeo_ch08= {1: 'Obra social (incluye PAMI)', 2: 'Mutual/Prepaga/Servicio de emergencia', 3:'Planes y seguros públicos', 4:'No paga ni le descuentan', 9:'Ns/Nr', 12:'Obra social y mutual / prepaga / servicio de emergencia', 13:'Obra social y planes y seguros públicos', 23:'Mutual / prepaga / servicio de emergencia / Planes y seguros públicos', 123: 'Obra social, mutual / prepaga / servicio de emergencia y planes y seguros públicos'}
mapeo_ch07= {1: 'Unido', 2: 'Casado', 3:'Separado o divorciado', 4:'Viudo', 5:'Soltero'}

# Aplicamos el mapeo a las columnas correspondientes
df_nuevo.loc[df_nuevo['ano4'] == 2024, 'ch04'] = df_nuevo.loc[df_nuevo['ano4'] == 2024, 'ch04'].map(mapeo_ch04)
df_nuevo.loc[df_nuevo['ano4'] == 2024, 'nivel_ed'] = df_nuevo.loc[df_nuevo['ano4'] == 2024, 'nivel_ed'].map(mapeo_nivel_ed)
df_nuevo.loc[df_nuevo['ano4'] == 2024, 'estado'] = df_nuevo.loc[df_nuevo['ano4'] == 2024, 'estado'].map(mapeo_estado)
df_nuevo.loc[df_nuevo['ano4'] == 2024, 'aglomerado'] = df_nuevo.loc[df_nuevo['ano4'] == 2024, 'aglomerado'].map(mapeo_aglomerado)
df_nuevo.loc[df_nuevo['ano4'] == 2024, 'cat_inac'] = df_nuevo.loc[df_nuevo['ano4'] == 2024, 'cat_inac'].map(mapeo_cat_inac)
df_nuevo.loc[df_nuevo['ano4'] == 2024, 'ch08'] = df_nuevo.loc[df_nuevo['ano4'] == 2024, 'ch08'].map(mapeo_ch08)
df_nuevo.loc[df_nuevo['ano4'] == 2024, 'ch07'] = df_nuevo.loc[df_nuevo['ano4'] == 2024, 'ch07'].map(mapeo_ch07)

# Reemplazamos el valor 'Entrevista individual no realizada (no respuesta al cuestion' por 'Entrevista individual no realizada (no respuesta al cuestionario individual)'
df_nuevo['estado'] = df_nuevo['estado'].replace('Entrevista individual no realizada (no respuesta al cuestion', 'Entrevista individual no realizada (no respuesta al cuestionario individual)')

# Cambiamos los NaN de la columna 'cat_inac' del año 2024 por 0.0
df_nuevo.loc[df_nuevo['ano4'] == 2024, 'cat_inac'] = df_nuevo.loc[df_nuevo['ano4'] == 2024, 'cat_inac'].fillna(0.0)

# Mostramos el DataFrame para verificar los cambios
df_nuevo.head()



```

```python
#queremos ver que observaciones sacamos, mirando primero la columna ch06 (edad)
valores_unicos = df_nuevo['ch06'].unique()
print(valores_unicos)

#sacamos las observaciones cateogoricas encontradas
df_nuevo = df_nuevo[~df_nuevo['ch06'].isin(['Menos de 1 año', '98 y más años'])]

#ahora sacamos las observaciones negativas de edad (ch06)
df_nuevo = df_nuevo[df_nuevo['ch06'] >= 0]

# Verificamos si hay alguna observación mayor a 100 en la columna 'ch06'
observaciones_mayores_100 = df_nuevo[df_nuevo['ch06'] > 100]
print(observaciones_mayores_100)

# Si hay observaciones mayores a 100, las mostramos
if not observaciones_mayores_100.empty:
    print("Observaciones con 'ch06' mayor a 100:")
    print(observaciones_mayores_100)
else:
    print("No hay observaciones con 'ch06' mayor a 100.")

#verificamos que no hay mayores a 100

# Convertimos la columna 'ch06' a float y luego a int para eliminar los decimales
df_nuevo['ch06'] = df_nuevo['ch06'].astype(float).astype(int)

#fuimos columna por columna para verificar que no haya valores raros en estas, lo hicimos con:
#valores_unicos1 = df_nuevo['...'].unique()
#print(valores_unicos1)

#ahora, vamos con la columna de ingresos (ipcf)
valores_unicos2 = df_nuevo['ipcf'].unique()
print(valores_unicos2)

# Verificamos si hay observaciones categóricas en la columna 'ipcf'
observaciones_categoricas_ipcf = df_nuevo[~df_nuevo['ipcf'].apply(lambda x: isinstance(x, (int, float)))]

# Si hay observaciones categóricas, las mostramos
if not observaciones_categoricas_ipcf.empty:
    print("Observaciones categóricas en 'ipcf':")
    print(observaciones_categoricas_ipcf)
else:
    print("No hay observaciones categóricas en 'ipcf'.")

#nos dio como resultado que no hay observaciones categoricas en ipcf

# Convertimos la columna 'ipcf' a float y luego a int para eliminar los decimales
df_nuevo['ipcf'] = df_nuevo['ipcf'].astype(float).astype(int)

# Mostramos el DataFrame para verificar los cambios
df_nuevo.head()

# Verificamos si hay valores negativos en la columna 'ipcf'
valores_negativos_ipcf = df_nuevo[df_nuevo['ipcf'] < 0]

# Si hay valores negativos, los eliminamos
if not valores_negativos_ipcf.empty:
    print("Observaciones con valores negativos en 'ipcf':")
    print(valores_negativos_ipcf)
    df_nuevo = df_nuevo[df_nuevo['ipcf'] >= 0]
else:
    print("No hay observaciones con valores negativos en 'ipcf'.")

#nos dio como resultado que no hay observaciones negativas en ipcf

# Mostramos el DataFrame para verificar los cambios
df_nuevo.head()
```

```python
#D
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

# Filtramos los datos para los años 2004 y 2024
df_filtrado = df_nuevo[df_nuevo['ano4'].isin([2004, 2024])]

# Seleccionamos las columnas especificadas
variables = ['ch04', 'ch06', 'ch07', 'ch08', 'nivel_ed', 'estado', 'cat_inac', 'ipcf']

# Convertimos las columnas categóricas a numéricas para la correlación
df_encoded = df_filtrado[variables].apply(lambda x: pd.factorize(x)[0])

# Calculamos la matriz de correlación
corr_matrix = df_encoded.corr()

# Creamos la cuadrícula y configuramos el gráfico principal
plot_grid = plt.GridSpec(1, 15, hspace=0.2, wspace=0.1)  # Configurarmos una cuadrícula de 1x15
fig = plt.figure(figsize=(15, 8))

# Gráfico de la matriz de correlación en la parte izquierda de la cuadrícula
ax_main = plt.subplot(plot_grid[:, :-1])  # Usamos las primeras 14 columnas de la cuadrícula
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', ax=ax_main)
ax_main.set_title('Matriz de Correlación para 2004 y 2024')

# Agregamos la barra de colores en la parte derecha del gráfico
ax_color = plt.subplot(plot_grid[:, -1])  # Usamos la última columna de la cuadrícula
palette = sns.color_palette("coolwarm", as_cmap=True)
color_min, color_max = -1, 1
n_colors = 256
bar_y = np.linspace(color_min, color_max, n_colors)
bar_height = bar_y[1] - bar_y[0]
ax_color.barh(
    y=bar_y,
    width=[5] * n_colors,
    left=[0] * n_colors,
    height=bar_height,
    color=palette(bar_y),
    linewidth=0
)

# Configuramos las etiquetas de la barra de colores
ax_color.set_xlim(1, 2)
ax_color.grid(False)
ax_color.set_facecolor('white')
ax_color.set_xticks([])  # Quitamos las etiquetas del eje x
ax_color.set_yticks(np.linspace(color_min, color_max, 5))
ax_color.set_yticklabels([f"{x:.1f}" for x in np.linspace(color_min, color_max, 5)], fontsize=12)
ax_color.yaxis.tick_right()  # Mostramos las etiquetas del eje y a la derecha

plt.show()
```
