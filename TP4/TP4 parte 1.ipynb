```python
# Convertir los nombres de las columnas a minúsculas para data2004
data2004.columns = data2004.columns.str.lower()

# Convertir los nombres de las columnas a minúsculas para data2024
data2024.columns = data2024.columns.str.lower()

# Mostrar las primeras filas de los DataFrames para verificar los cambios
data2004.head(), data2024.head()
```

```python
# Unir ambos DataFrames en una sola base (uno debajo del otro)
df_combined = pd.concat([data2004, data2024], ignore_index=True)

# Mostrar el DataFrame combinado para verificar
df_combined
```

```python
# Filtrar los datos de df_combined para que solo queden los pertenecientes a Partidos del GBA y Ciudad de Buenos Aires
df_bsas = df_combined[df_combined['aglomerado'].isin([32, 33, 'Partidos del GBA','Ciudad de Buenos Aires'])]
#sabemos que eran 32 y 33 por el diccionario de amos años
# Mostrar las primeras filas del DataFrame filtrado para verificar
df_bsas
```

```python
#1. Porcentaje de ingresos provenientes de personas ocupadas en el hogar
# Crear una columna binaria para identificar si el estado es 'Ocupado'
df_selec['es_ocupado'] = (df_selec['estado'] == 'Ocupado').astype(int)

# Calcular el ingreso total generado por personas ocupadas en el hogar
df_selec['ingreso_ocupados'] = df_selec['es_ocupado'] * df_selec['ipcf']  # IPCF = ingreso per cápita familiar

# Calcular el ingreso total del hogar
df_selec['ingreso_total_hogar'] = df_selec.groupby('nro_hogar')['ipcf'].transform('sum')

# Calcular el porcentaje de ingresos generados por personas ocupadas
df_selec['porcentaje_ingreso_ocupados'] = df_selec.groupby('nro_hogar')['ingreso_ocupados'].transform('sum') / df_selec['ingreso_total_hogar']

#2. Nivel educativo promedio del hogar
# Mapear niveles educativos a valores numéricos #lo hacemos de forma jerarquica, no como aparece en el diccionario
nivel_educativo_map = {
    'Sin instrucción': 0,
    'Primaria Incompleta (incluye educación especial)': 1,
    'Primaria Completa': 2,
    'Secundaria Incompleta': 3,
    'Secundaria Completa': 4,
    'Superior Universitaria Incompleta': 5,
    'Superior Universitaria Completa': 6,
    'Ns./Nr.': None
}

# Convertir a valores numéricos
df_selec['nivel_ed_num'] = df_selec['nivel_ed'].map(nivel_educativo_map)

# Calcular el nivel educativo promedio por hogar
df_selec['nivel_educativo_promedio'] = df_selec.groupby('nro_hogar')['nivel_ed_num'].transform('mean')

#3. Relacion entre edad y estado laboral
# Categorizar edades en grupos de riesgo
def categorizar_edad(edad):
    if edad < 18:
        return 'Menor de edad'
    elif 18 <= edad <= 24:
        return 'Joven'
    elif 25 <= edad <= 54:
        return 'Adulto'
    else:
        return 'Mayor'

df_selec['grupo_edad'] = df_selec['ch06'].apply(categorizar_edad)


#vemos como quedo df_selec
df_selec



```
