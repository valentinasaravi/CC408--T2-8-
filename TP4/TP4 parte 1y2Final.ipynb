```python
# Convertir los nombres de las columnas a minúsculas para data2004
data2004.columns = data2004.columns.str.lower()

# Convertir los nombres de las columnas a minúsculas para data2024
data2024.columns = data2024.columns.str.lower()

# Mostrar las primeras filas de los DataFrames para verificar los cambios
data2004.head(), data2024.head()
```

```python
# Unir ambos DataFrames en una sola base (uno debajo del otro)
df_combined = pd.concat([data2004, data2024], ignore_index=True)

# Mostrar el DataFrame combinado para verificar
df_combined
```

```python
# Filtrar los datos de df_combined para que solo queden los pertenecientes a Partidos del GBA y Ciudad de Buenos Aires
df_bsas = df_combined[df_combined['aglomerado'].isin([32, 33, 'Partidos del GBA','Ciudad de Buenos Aires'])]
#sabemos que eran 32 y 33 por el diccionario de amos años
# Mostrar las primeras filas del DataFrame filtrado para verificar
df_bsas
```

```python
#voy a combinar mi base de datos de la encuesta de hogar con la base de la encuesta individual de cada año (usada en el TP3)
# Cargar la base de datos CSV
#file_path_csv = '/Users/valentinasaravi/Documents/Ciencia de datos/Tutoriales/TPs/TP3/df_filtered.csv'
#file_path_csv = 'C:/Users/matug/OneDrive/Documentos/UdeSA/Ciencia de datos/TP4/Bases de datos TP3/df_filtered.csv'
file_path_csv = 'C:/Users/Usuario/Documents/Ciencia de datos/Enviroment/TP4/df_filtered.csv'
data_tp3 = pd.read_csv(file_path_csv)

# Realizar el merge utilizando las variables 'codusu' y 'nro_hogar'
df_combined_final = pd.merge(df_bsas, data_tp3, on=['codusu', 'nro_hogar'], how='inner')
df_combined_final
print("Dimensiones de df_combined_final:", df_combined_final.shape)
# Identificar columnas que terminan en _x y _y
columnas_x = [col for col in df_combined_final.columns if col.endswith('_x')]
columnas_y = [col for col in df_combined_final.columns if col.endswith('_y')]

# Mostrar las columnas encontradas
print("Columnas que terminan con '_x':")
print(columnas_x)

print("\nColumnas que terminan con '_y':")
print(columnas_y)

# Listas de columnas con sufijos _x y _y
columnas_x = ['ano4_x', 'trimestre_x', 'region_x', 'mas_500_x', 'aglomerado_x', 'pondera_x', 
              'itf_x', 'decifr_x', 'idecifr_x', 'rdecifr_x', 'gdecifr_x', 'pdecifr_x', 
              'adecifr_x', 'ipcf_x', 'deccfr_x', 'ideccfr_x', 'rdeccfr_x', 'gdeccfr_x', 
              'pdeccfr_x', 'adeccfr_x', 'pondih_x']

columnas_y = ['ano4_y', 'trimestre_y', 'region_y', 'mas_500_y', 'aglomerado_y', 'pondera_y', 
              'itf_y', 'decifr_y', 'idecifr_y', 'rdecifr_y', 'gdecifr_y', 'pdecifr_y', 
              'adecifr_y', 'ipcf_y', 'deccfr_y', 'ideccfr_y', 'rdeccfr_y', 'gdeccfr_y', 
              'pdeccfr_y', 'adeccfr_y', 'pondih_y']

# Comparar columnas _x y _y de forma explícita para asegurar que todas sean mencionadas
resultado_comparacion = {}

for col_x, col_y in zip(columnas_x, columnas_y):
    if col_x in df_combined_final.columns and col_y in df_combined_final.columns:
        # Comparar si los datos son iguales (considerando valores NaN como equivalentes)
        iguales = df_combined_final[col_x].fillna('MISSING').equals(df_combined_final[col_y].fillna('MISSING'))
        resultado_comparacion[f"{col_x} vs {col_y}"] = iguales
    else:
        # Si no existen, indicarlo explícitamente
        resultado_comparacion[f"{col_x} vs {col_y}"] = "Una o ambas columnas no existen"

# Mostrar los resultados de forma detallada
print("Resultados de la comparación explícita:")
for relacion, resultado in resultado_comparacion.items():
    print(f"{relacion}: {resultado}")

# Lista de pares de columnas específicas a comparar (las que dicen False)
pares_columnas = [
    ("trimestre_x", "trimestre_y"),
    ("region_x", "region_y"),
    ("aglomerado_x", "aglomerado_y"),
    ("decifr_x", "decifr_y"),
    ("rdecifr_x", "rdecifr_y"),
    ("gdecifr_x", "gdecifr_y"),
    ("adecifr_x", "adecifr_y"),
    ("ipcf_x", "ipcf_y"),
    ("deccfr_x", "deccfr_y"),
    ("rdeccfr_x", "rdeccfr_y"),
    ("gdeccfr_x", "gdeccfr_y"),
    ("adeccfr_x", "adeccfr_y")
]

```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Generar los valores de λ
lambda_values = 10 ** np.linspace(-5, 5, 11)  # Más granularidad en los valores de λ
C_values = [1 / lam for lam in lambda_values]  # Convertir λ a C

# Almacenar los resultados
results = []
penalty_types = ['l1', 'l2']
ignored_proportions = []  # Para guardar las proporciones de coeficientes cero para LASSO

for penalty in penalty_types:
    print(f"\nProcesando penalidad: {penalty.upper()}")
    for C, lam in zip(C_values, lambda_values):
        logistic = LogisticRegression(penalty=penalty, solver='liblinear', C=C, max_iter=1000)
        
        # Calcular el MSE en validación cruzada
        scores = cross_val_score(logistic, X_train_04_transformed, y_train_04, cv=10, scoring='accuracy')
        mse_values = 1 - scores  # Convertir Accuracy a MSE
        
        # Almacenar resultados
        for mse in mse_values:
            results.append({'Lambda': lam, 'Penalidad': penalty.upper(), 'MSE': mse})
        
        # Calcular proporción de coeficientes cero para LASSO
        if penalty == 'l1':
            logistic.fit(X_train_04_transformed, y_train_04)  # Ajustar el modelo completo
            coef_zeros = (logistic.coef_ == 0).sum()  # Contar coeficientes iguales a cero
            total_coefs = logistic.coef_.size  # Total de coeficientes
            ignored_proportions.append({'Lambda': lam, 'Proporción Ignorada': coef_zeros / total_coefs})

# Crear DataFrame para los resultados generales
results_df = pd.DataFrame(results)

# Filtrar resultados para L2 y generar el boxplot
results_l2 = results_df[results_df['Penalidad'] == 'L2']



```

```python
# Identificar los mejores lambdas (menor MSE) para L1 y L2
best_lambda_l1 = results_df[results_df['Penalidad'] == 'L1'].groupby('Lambda')['MSE'].mean().idxmin()
best_lambda_l2 = results_df[results_df['Penalidad'] == 'L2'].groupby('Lambda')['MSE'].mean().idxmin()

print(f"Mejor lambda para L1 (LASSO): {best_lambda_l1}")
print(f"Mejor lambda para L2 (Ridge): {best_lambda_l2}")

# Entrenar modelos con los mejores lambdas para obtener los coeficientes
# Para L1 (LASSO)
logistic_l1 = LogisticRegression(penalty='l1', solver='liblinear', C=1/best_lambda_l1, max_iter=1000)
logistic_l1.fit(X_train_04_transformed, y_train_04)
coefficients_l1 = logistic_l1.coef_

# Para L2 (Ridge)
logistic_l2 = LogisticRegression(penalty='l2', solver='liblinear', C=1/best_lambda_l2, max_iter=1000)
logistic_l2.fit(X_train_04_transformed, y_train_04)
coefficients_l2 = logistic_l2.coef_

# Imprimir coeficientes
print("\nCoeficientes para L1 (LASSO):")
print(coefficients_l1)

print("\nCoeficientes para L2 (Ridge):")
print(coefficients_l2)
```

```python
# Convertir Lambda a categórico
results_l2['Lambda_Cat'] = results_l2['Lambda'].apply(lambda x: f"$10^{{{int(np.log10(x))}}}$")

# Graficar con Seaborn
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=results_l2,
    x='Lambda_Cat',
    y='MSE',
    palette='pastel'
)
plt.xlabel("Lambda (Escala Logarítmica)")
plt.ylabel("MSE")
plt.title("Distribución del Error de Predicción (MSE) para Ridge (L2)")
plt.tight_layout()
plt.show()

# Crear DataFrame para las proporciones ignoradas en LASSO
ignored_proportions_df = pd.DataFrame(ignored_proportions)

# Graficar el promedio de la proporción de coeficientes ignorados en función de λ
plt.figure(figsize=(10, 6))
sns.lineplot(
    data=ignored_proportions_df, 
    x='Lambda', 
    y='Proporción Ignorada', 
    marker='o'
)
plt.xscale('log')  # Escala logarítmica para λ
plt.xlabel("Lambda (Escala Logarítmica)")
plt.ylabel("Proporción de Coeficientes Ignorados")
plt.title("Proporción de Variables Ignoradas por LASSO (L1)")
plt.tight_layout()
plt.show()
```
